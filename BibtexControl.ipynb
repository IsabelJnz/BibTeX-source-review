```python
# remove ä,ü,ö
# this function splits the BibTeX file into seperate BibTeX items
def clean_sources(d):
    d=d.replace('ü', '{\"u}')
    d=d.replace('ä', '{\"a}')
    d=d.replace('ö', '{\"o}')
    d=d.replace('Ü', '{\"U}')
    d=d.replace('Ä', '{\"A}')
    d=d.replace('Ö', '{\"O}')
    d=d.replace('\ss', '{\ß}')

    
    sc=d.split('@')

    sources=[]

    for x in range(0,len(sc)-1):
        tmp=sc[x]
        sources.append(f'@{tmp}')
    return sources
```

```python
# this function splits the BibTeX file into seperate BibTeX items
def subsequent_processing(a):
    sca=a.split('@')

    src=[]

    for x in range(1,len(sca)-1):
        tmp=sca[x]
        src.append(f'@{tmp}')
    
    return(src)
```

```python
#get warings for violating the rules of the datestamp
def get_date(sources):
  sources_final=[]

  for x in range(0,len(sources)-1):
    tmp=sources[x]

    strtmp="".join(tmp)
    strtmp=strtmp.lower()

    y=strtmp.find('year')
    
    sources_final.append(tmp) 

    if y!=-1:
      sk1=strtmp.find('{', y)
      sk2=strtmp.find('}', sk1)
      sum=sk2-sk1
      
      if sum > 5:
        del sources_final[-1]
        sources_final.append(f'\n{tmp} |----> WARNING: Please adapt datestamp above -> (year only)! ---- \n')
      

  return sources_final
```

```python
#get warings for violating the rules of the abbreviations
def get_abbreviation(sources):
  sources_final=[]

  for x in range(0,len(sources)-1):
    tmp=sources[x]

    strtmp="".join(tmp)
    strtmp=strtmp.lower()

    i=strtmp.find('@inproceedings')
    a=strtmp.find('@article')
    
    sources_final.append(tmp) 

  	  
    if i!=-1 or a!=-1:
      j=strtmp.find('journal')
      gk=strtmp.find(',', j)
      k=strtmp.find('(', j, gk)

      if j!=-1 and k!=-1:
        del sources_final[-1]
        sources_final.append(f'\n{tmp} |----> WARNING: Please remove abbreviation in journal above! ---- \n')


  return sources_final
```

```python
#get warings for violating the rules of the doi specification
def get_doi(sources):
  sources_final=[]

  for x in range(0,len(sources)-1):
    tmp=sources[x]

    strtmp="".join(tmp)
    strtmp=strtmp.lower()

    i=strtmp.find('@inproceedings')
    a=strtmp.find('@article')
    
    sources_final.append(tmp) 

    if i!=-1 or a!=-1:
      j=strtmp.find('journal')
      gk=strtmp.find(',', j)
      k=strtmp.find('(', j, gk)
      d=strtmp.find('doi')

      
      if d==-1:
        del sources_final[-1]
        sources_final.append(f'\n{tmp} |----> WARNING: Please add DOI of the article above! ----\n')

      

  return sources_final
```

```python
#get warings for violating the rules of the location specification
def get_location(sources):
  sources_final=[]

  for x in range(0,len(sources)-1):
    tmp=sources[x]

    strtmp="".join(tmp)
    strtmp=strtmp.lower()

    b=strtmp.rfind('@book')
    i=strtmp.find('@inproceedings')
    
    sources_final.append(tmp) 


    if b!=-1 or i!=-1:
      ad=strtmp.find('address', b, len(strtmp))

      if ad==-1:
        del sources_final[-1]
        sources_final.append(f'\n{tmp} |----> WARNING: Please add location above -> (Street, State, Country)! ----\n')
      

  return sources_final
```

```python
#get warings for violating the rules of the page specification
def get_pages(sources):
  sources_final=[]

  for x in range(0,len(sources)-1):
    tmp=sources[x]

    strtmp="".join(tmp)
    strtmp=strtmp.lower()    

    p=strtmp.find('pages')

    sources_final.append(tmp) 


    if p==-1:
      del sources_final[-1]
      sources_final.append(f'\n{tmp} |----> WARNING: Please add the information about the number of pages! ---- \n')


  return sources_final
```

```python
#get warings for violating the rules of the author specification
def get_authors(sources):
  sources_final=[]

  for x in range(0,len(sources)-1):
    tmp=sources[x]

    strtmp="".join(tmp)
    strtmp=strtmp.lower()

    at=strtmp.find('author')
    
    sources_final.append(tmp) 

  	  
    if at!=-1:
      gk=strtmp.find(',', at)
      na=strtmp.count('and', at, gk)


      if na>3:
        del sources_final[-1]
        sources_final.append(f'\n{tmp} |----> WARNING: Please adapt author list. -> (max three authors with name, more then three authors please add and others after the third author)! ---- \n')


  return sources_final
```

```python
#get warings for violating the rules of the formular specification
def get_formular(sources):
  sources_final=[]

  for x in range(0,len(sources)-1):
    tmp=sources[x]

    strtmp="".join(tmp)
    strtmp=strtmp.lower()

    sqr=strtmp.find('sqr(')
    l=strtmp.find('_')
    r=strtmp.find('^')
    e=strtmp.find('=')
    p=strtmp.find('\%')
    a=strtmp.find('\&')

    
    sources_final.append(tmp) 

  	  
    if sqr!=-1 and l!=-1 and r!=-1 and e!=-1 and p!=-1 and a!=-1:
      del sources_final[-1]
      sources_final.append(f'\n{tmp} |----> WARNING: Please adapt formular $...$ are essentiell! -> (https://en.wikibooks.org/wiki/LaTeX/Mathematics)! ---- \n')


  return sources_final
```

```python
#get warings for violating the rules of the language distinction
def get_german_language(sources):
  sources_final=[]

  for x in range(0,len(sources)-1):
    tmp=sources[x]

    strtmp="".join(tmp)
    strtmp=strtmp.lower()

    #de=strtmp.find(' der ')
    di=strtmp.find(' die ')
    da=strtmp.find(' das ')
    se=strtmp.find(' sein ')
    un=strtmp.find(' und ')
    ei=strtmp.find(' ein ')
    zu=strtmp.find(' zu ')
    vo=strtmp.find(' von ')
    ha=strtmp.find(' haben ')
    we=strtmp.find(' werden ')
    mi=strtmp.find(' mit ')
    fu=strtmp.find(' f{\"u}r ')
    au=strtmp.find(' auf ')
    si=strtmp.find(' sich ')
    es=strtmp.find(' es ')
    ni=strtmp.find(' nicht ')
    er=strtmp.find(' er ')
    auc=strtmp.find(' auch ')
    al=strtmp.find(' als ')

    
    sources_final.append(tmp) 

    if di!=-1 or da!=-1 or se!=-1 or un!=-1 or ei!=-1 or zu!=-1 or vo!=-1 or ha!=-1 or we!=-1 or mi!=-1 or fu!=-1 or au!=-1 or si!=-1 or es!=-1 or ni!=-1 or er!=-1 or auc!=-1 or al!=-1:  	  
        del sources_final[-1]
        sources_final.append(f'\n{tmp} |----> WARNING: Please check the BibTeX format for german sources!' +  '-> (instead of: {title, etc.}, use: {{title, etc.}})! ---- \n')


  return sources_final
```

```python
#get warings for violating the rules of the edition specification
def get_edition(sources):
  sources_final=[]

  for x in range(0,len(sources)-1):
    tmp=sources[x]

    strtmp="".join(tmp)
    strtmp=strtmp.lower()

    e=strtmp.find('edition')
    
    sources_final.append(tmp) 

  	  
    if e!=-1:
      gk=strtmp.find(',', e)
      st=strtmp.find('st', e, gk)
      nd=strtmp.find('nd', e, gk)
      rd=strtmp.find('rd', e, gk)
      th=strtmp.find('th', e, gk)

      if st==-1 and nd==-1 and rd==-1 and th==-1:
        del sources_final[-1]
        sources_final.append(f'\n{tmp} |----> WARNING: Please adapt edition -> (add: st, nd, rd or th!) ---- \n')


  return sources_final
```

```python
#get warings for violating the rules of the url specification
def get_url(sources):
  sources_final=[]

  for x in range(0,len(sources)-1):
    tmp=sources[x]

    strtmp="".join(tmp)
    strtmp=strtmp.lower()

    e=strtmp.find('url')
    
    sources_final.append(tmp) 

  	  
    if e!=-1:
        del sources_final[-1]
        sources_final.append(f'\n{tmp} |----> WARNING: Please remove URL! ---- \n')


  return sources_final
```
